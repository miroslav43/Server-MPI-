```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

```python
df = pd.read_csv("output/tasks.csv")


# Print a summary of the data
print("Data Summary:")
print(df.describe())
print("\nFirst few rows:")
print(df.head())
```

```python

plt.figure(figsize=(10, 6))
plt.bar(df['client_id'], df['total_time'], color='steelblue')
plt.xlabel('Client ID')
plt.ylabel('Total Time (s)')
plt.title('Total Execution Time per Command')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('results/plot_total_time_per_command.png')
plt.show()
```

```python
# 2. Average time per command type
avg_time_by_command = df.groupby('command')['total_time'].mean()

plt.figure(figsize=(8,5))
avg_time_by_command.plot(kind='bar', color='orange')
plt.xlabel('Command Type')
plt.ylabel('Average Total Time (s)')
plt.title('Average Total Time by Command Type')
plt.tight_layout()
plt.savefig('results/plot_average_time_by_command_type.png')
plt.show()
```

```python
# 3. Histogram of total times
# This shows how many commands fall into different ranges of total time
plt.figure(figsize=(8,5))
plt.hist(df['total_time'], bins=10, color='green', edgecolor='black')
plt.xlabel('Total Time (s)')
plt.ylabel('Number of Commands')
plt.title('Distribution of Total Times')
plt.tight_layout()
plt.savefig('results/plot_histogram_total_times.png')
plt.show()
```

```python


# 3. Box plot of total_time grouped by command type
# Group the dataframe by command type
commands_grouped = df.groupby('command')
# Create a box plot of total_time by command type
plt.figure(figsize=(10, 6))
df.boxplot(column='total_time', by='command', grid=True)
plt.xlabel('Command Type')
plt.ylabel('Total Time (s)')
plt.title('Distribution of Total Time by Command Type')
plt.suptitle('')  # Removes the default subtitle
plt.tight_layout()
plt.savefig('results/plot_boxplot_by_command.png')
plt.show()
```
